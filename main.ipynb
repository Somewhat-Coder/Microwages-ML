Visualizing Outliers

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))

sns.boxplot(df['puma'], flierprops={"marker": "x"}, ax=axes[0, 0])
axes[0, 0].set_title('Puma')

sns.boxplot(df['perwt'], flierprops={"marker": "x"}, ax=axes[0, 1])
axes[0, 1].set_title('Perwt')

sns.boxplot(df['age'], flierprops={"marker": "x"}, ax=axes[0, 2])
axes[0, 2].set_title('Age')

sns.boxplot(df['edyrs'], flierprops={"marker": "x"}, ax=axes[1, 0])
axes[1, 0].set_title('Edyrs')

sns.boxplot(df['expyrs'], flierprops={"marker": "x"}, ax=axes[1, 1])
axes[1, 1].set_title('Expyrs')

sns.boxplot(df['wkwage'], flierprops={"marker": "x"}, ax=axes[1, 2])
axes[1, 2].set_title('Wkwage')

plt.tight_layout()

plt.show()
```

Visualizing Categorical Variables

```python
fig2, axes2 = plt.subplots(2, 2, figsize=(20, 10))

region_counts = dict(df['region'].value_counts())
region_counts = dict(sorted(region_counts.items()))

palette = ["#fee090", "#fdae61", "#4575b4", "#313695", "#e0f3f8", "#abd9e9", "#d73027", "#a50026"]

sns.set_theme(rc = {'figure.figsize':(8, 5)})
sns.barplot(data=region_counts, palette=palette, ax=axes2[0, 0])
axes2[0, 0].set_xlabel("Regions")
axes2[0, 0].set_ylabel('COunt')

for bars in axes2[0, 0].containers:
    axes2[0, 0].bar_label(bars)

colors = sns.color_palette('pastel')[0:5]
sns.countplot(data = df, x = 'race_nonwhite', palette=colors, hue='race_nonwhite', legend=False, ax=axes2[0, 1])
axes2[0, 1].set_xticks([0,1])
axes2[0, 1].set_xticklabels(['White', 'Black'])
axes2[0, 1].set_xlabel("Race")

for bars in axes2[0, 1].containers:
    axes2[0, 1].bar_label(bars)

sns.countplot(data = df, x = 'female', palette=['lightblue', 'orange'], hue='female', legend=False,  ax=axes2[1, 0])

axes2[1, 0].set_xticks([0,1])
axes2[1, 0].set_xticklabels(['Male', 'Female'])
axes2[1, 0].set_xlabel("Gender")

for bars in axes2[1, 0].containers:
    axes2[1, 0].bar_label(bars)

sns.countplot(data = df, x = 'statefip', palette=colors, hue='statefip', legend=False,  ax=axes2[1, 1])

axes2[1, 1].set_xlabel("Statefip")
axes2[1, 1].set_xticklabels(axes2[1, 1].get_xticklabels(), rotation=90)
```

Visualizing Feature Relationships

```python
sns.set_theme(rc = {'figure.figsize':(8, 5)})
ax = sns.scatterplot(data=df, x=df['edyrs'], y=df['wkwage'], hue='wkwage')

ax.set_xlabel("Education in years")
ax.set_ylabel("Weekly wage")
ax.set_title("Education in years and wage")
```

```python
edyrs_wkwage_mean = df.groupby('edyrs')['wkwage'].mean().reset_index()

sns.set_theme(rc = {'figure.figsize':(8, 5)})
ax = sns.scatterplot(data=edyrs_wkwage_mean, x=df['edyrs'], y=df['wkwage'], hue='wkwage')

ax.set_xlabel("Education in years")
ax.set_ylabel("Average Weekly wage")
ax.set_title("Education in years and average wage")
```

```python
df['expyrs_rounded'] = df['expyrs'].round()

exp_wage = df.groupby('expyrs_rounded')['wkwage'].mean().reset_index()

sns.set_theme(rc = {'figure.figsize':(8, 5)})
sns.lineplot(x=df['expyrs_rounded'], y=df['wkwage'])

ax.set_xlabel("Experience in years")
ax.set_ylabel("Average Weekly wage")
ax.set_title('Experience and wage')
```

```python
selected_columns = df[['statefip','wkwage', 'age']]

heatmap_data = selected_columns.pivot_table(index='age', columns='statefip', values='wkwage', aggfunc='mean')

sns.set_theme(rc = {'figure.figsize':(8, 5)})
sns.heatmap(heatmap_data, cmap='viridis', fmt='.2f', cbar_kws={'label': 'Average Wage'})
plt.title('Heatmap of Average Wage by Age and State')
plt.show()
```

```python
result = df.groupby(['region', 'female'])['wkwage'].agg(['min', 'max']).reset_index()

for index, row in result.iterrows():
    duplicated_row = row.copy()  # Create a copy of the current row
    duplicated_row['min'] = row['max']  # Move the value to the existing column
    result = result._append(duplicated_row, ignore_index=True)

result = result.rename(columns={'min': 'min/max'})
result = result.drop('max', axis=1)

sns.set_theme(rc = {'figure.figsize':(8, 5)})
ax = sns.barplot(data=result, x="region",y='min/max', hue="female")
ax.set_ylabel("Min / Max wage")
ax.set_title("Salaries of Males and Females in their region")
```

```python
result = df[['edyrs','expyrs','wkwage']]

result.iloc[:, 0] = result['edyrs'].round().astype('int')
result.iloc[:, 1] = result['expyrs'].round().astype('int')

heatmap_data = result.pivot_table(index='edyrs', columns='expyrs', values='wkwage', aggfunc='mean')

sns.set_theme(rc = {'figure.figsize':(8, 5)})
sns.heatmap(heatmap_data, cmap='viridis', fmt='.2f', cbar_kws={'label': 'Average Wage'})
plt.title('Heatmap of Average Wage by Edyrs and Expyrs')
plt.show()
```

```python
edyrs_wkwage_mean = df.groupby('edyrs')['wkwage'].mean().reset_index()

sns.set_theme(rc = {'figure.figsize':(20, 10)})

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(df['perwt'], df['edyrs'], df['wkwage'], c='cyan')

ax.set_xlabel('perwt')
ax.set_ylabel('edyrs')
ax.set_zlabel('Wkwage')
ax.set_title('Perwt, edyrs and Wkwage')

plt.show()
```

```python

occu_data = pd.DataFrame(columns=['occupation', 'expyrs_rounded', 'wkwage'])

cols = ['occ_managprof', 'occ_techsalad', 'occ_service', 'occ_farm', 'occ_product', 
                    'occ_operator', 'occ_service_np', 'occ_service_broad', 'expyrs_rounded']

a = df.groupby(cols)['wkwage'].mean().reset_index()

for col in cols[:8]:
    result = a.loc[a[col] == 1, [col, 'expyrs_rounded', 'wkwage']]
    result[col] = result[col].replace({1: col})
    result = result.rename(columns={col: 'occupation'})
    occu_data = pd.concat([occu_data, result], ignore_index=True)

ax = sns.lineplot(occu_data, x="expyrs_rounded", hue="occupation", y="wkwage")
ax.set_xlabel('Experience in years')
ax.set_ylabel('Average weekly wage')
```
